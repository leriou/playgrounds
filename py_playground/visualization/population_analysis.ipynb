# 人口数据分析

暂时先使用2010年的数据进行分析后，，替换为2020年的人口普查数据


```python
import pandas as pd

base_year = 2010
peoples = pd.read_csv("../A0301a.csv")

def age_process(s):
    if s.isnumeric():
        return int(s)
    else:
        return None

def born_cal(n):
    return base_year - n
"""
处理数据
"""
peoples['age'] = peoples['年龄'].apply(age_process)
peoples['born'] = peoples['age'].apply(born_cal)

data = peoples[peoples['age'] >= 0]

# 医药
medical_period = 60
data['medical'] = data['born'] + medical_period

# 住房
house = 31
big_house = 42
last_house = 65

data['l_house'] = data['born'] + house
data['m_house'] = data['born'] + big_house
data['b_house'] = data['born'] + last_house

# 消费 46-48
consume = 47
data['consume'] = data['born'] + consume

# 奢侈品
luxury = 52 
data['consume'] = data['born'] + luxury
```

```python
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'svg'

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(data['born'], data['合计'])
ax.plot(data['consume'], data['合计'])
ax.plot(data['medical'], data['合计'])

# ax.plot(data['b_house'], data['合计'])
# ax.plot(data['m_house'], data['合计'])
# ax.plot(data['l_house'], data['合计'])


plt.show()
```
